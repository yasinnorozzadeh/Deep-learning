```python
import wandb
import tensorflow as tf
from sklearn.model_selection import train_test_split
from tensorflow.keras import datasets, optimizers, losses, models
from wandb.keras import WandbCallback
from tensorflow.keras import layers
```

```python
wandb.init(project='Cifar10_CNN')
```

```python
config = wandb.config
config.learning_rate = 0.001
```

```python
cifar10 = datasets.cifar10
(X_train, Y_train), (X_Test, Y_Test) = cifar10.load_data()

X_val, X_test, Y_val, Y_test = train_test_split(X_Test,Y_Test, shuffle=True, test_size=0.5)
X_train, X_test, X_val = X_train / 255.0, X_test / 255.0, X_val / 255.0
```

```python
model_CNN = tf.keras.models.Sequential([
  layers.Conv2D(32,(3,3), activation='relu', input_shape=(32,32,3)),
  layers.MaxPooling2D(),
  layers.Conv2D(64,(3,3), activation='relu'),
  layers.MaxPooling2D(),
  layers.Conv2D(64,(5,5), activation='relu'),

  layers.Flatten(),

  layers.Dense(100,activation='relu'),
  layers.Dense(10,activation='softmax')
])
```

```python
model_CNN.compile(optimizer = optimizers.Adam(learning_rate = config.learning_rate),
                  loss = losses.sparse_categorical_crossentropy,
                  metrics=["accuracy"])

model_CNN.fit(X_train, Y_train, epochs=10, validation_data=(X_val, Y_val), callbacks=[WandbCallback()])
```

```python
model_CNN.evaluate(X_test,Y_test)
```

```python
model_MLP = tf.keras.models.Sequential([
  tf.keras.layers.Flatten(input_shape=(32, 32, 3)),
  tf.keras.layers.Dense(128, activation='relu'),
  tf.keras.layers.Dropout(0.2),
  tf.keras.layers.Dense(10, activation='softmax')
])
```

```python
model_MLP.compile(optimizer=optimizers.Adam(learning_rate=config.learning_rate),
                  loss= losses.sparse_categorical_crossentropy,
                  metrics=['accuracy'])
```

```python
model_MLP.fit(X_train, Y_train, epochs=10)
```

```python
model_MLP.evaluate(X_test, Y_test)
```
