```python
import tensorflow as tf
import wandb
from wandb.keras import WandbCallback
from tensorflow.keras import models, losses, datasets, optimizers
from tensorflow.keras.layers import Dense, Conv2D, MaxPool2D, Flatten
from sklearn.model_selection import train_test_split
```

```python
wandb.init(project='Mnist')
```

```python
config=wandb.config
config.learning_rate=0.001
```

```python
(X_train, Y_train), (X_Test, Y_Test) = datasets.mnist.load_data()
X_val, X_test, Y_val, Y_test = train_test_split(X_Test,Y_Test, test_size=0.5)
X_train,X_val,X_test = X_train / 255.0,X_val/255.0,X_test / 255.0
```

```python
model_CNN = tf.keras.models.Sequential([
    Conv2D(32, (3,3), activation='relu', input_shape=(28,28,1)),
    MaxPool2D(),
    Conv2D(64, (3,3), activation='relu'),
    MaxPool2D(),
    Conv2D(64, (5,5), activation='relu'),

    Flatten(),

    Dense(784, activation='relu'),
    Dense(10, activation='softmax')
])
```

```python
model_CNN.compile(optimizer= optimizers.Adam(config.learning_rate),
              loss = tf.keras.losses.sparse_categorical_crossentropy,
              metrics=['accuracy'])

model_CNN.fit(X_train, Y_train, epochs=5 , validation_data=(X_val, Y_val),callbacks=[WandbCallback()])
```

```python
model_CNN.evaluate(X_test,Y_test)
```

```python
model_MLP = tf.keras.models.Sequential([
  tf.keras.layers.Flatten(input_shape=(28, 28)),
  tf.keras.layers.Dense(90, activation='relu'),
  tf.keras.layers.Dense(128, activation='relu'),
  tf.keras.layers.Dense(10, activation='softmax')
])
```

```python
model_MLP.compile(optimizer=optimizers.Adam(learning_rate=config.learning_rate),
                  loss= losses.sparse_categorical_crossentropy,
                  metrics=['accuracy'])
```

```python
model_MLP.fit(X_train, Y_train, epochs=5)
```

```python
model_MLP.evaluate(X_test, Y_test)
```
