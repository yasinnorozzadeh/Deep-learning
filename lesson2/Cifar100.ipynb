```python
import tensorflow as tf
import wandb
from tensorflow.keras import layers, optimizers, losses
from sklearn.model_selection import train_test_split
from wandb.keras import WandbCallback
```

```python
wandb.init(project="Cifar100")
```

```python
config = wandb.config
config.learning_rate = 0.001
```

```python
cifar_100 = tf.keras.datasets.cifar100
(X_train, Y_train), (X_test, Y_test) = cifar_100.load_data()

X_train, X_test = X_train / 255.0, X_test / 255.0
X_train, X_val, Y_train, Y_val = train_test_split(X_train, Y_train, test_size=0.2, shuffle=True)
```

```python
model_CNN = tf.keras.models.Sequential([
    layers.Conv2D(32, (3,3), activation='relu', input_shape=(32,32,3)),
    layers.MaxPooling2D(),
    layers.Conv2D(64, (3,3), activation='relu'),
    layers.MaxPooling2D(),
    layers.Conv2D(64, (5,5), activation='relu'),

    layers.Flatten(),

    tf.keras.layers.Dense(64, activation='relu'),
    tf.keras.layers.Dense(100, activation='softmax')
])
```

```python
model_CNN.compile(optimizer=tf.keras.optimizers.Adam(config.learning_rate),
              loss = tf.keras.losses.sparse_categorical_crossentropy,
              metrics=['accuracy'])

model_CNN.fit(X_train, Y_train, epochs=10 , validation_data=(X_val, Y_val),callbacks=[WandbCallback()])
```

```python
model_CNN.evaluate(X_test,Y_test)
```

```python
model_MLP = tf.keras.models.Sequential([
  tf.keras.layers.Flatten(input_shape=(32, 32, 3)),
  tf.keras.layers.Dense(128, activation='relu'),
  tf.keras.layers.Dense(64, activation='relu'),
  tf.keras.layers.Dropout(0.2),
  tf.keras.layers.Dense(100, activation='softmax')
])
```

```python
model_MLP.compile(optimizer= optimizers.Adam(learning_rate=config.learning_rate),
                  loss= losses.sparse_categorical_crossentropy,
                  metrics=['accuracy'])
```

```python
model_MLP.fit(X_train, Y_train, epochs=5)
```

```python
model_MLP.evaluate(X_test, Y_test)
```
